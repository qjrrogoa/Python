## 피벗테이블 사용하기

```python
#라이브러리 불러오기
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/도로교통공단_사망 교통사고 정보_20201231.csv',encoding='cp949')
```

```python
df.columns
```

```python
#2개 이상의 데이터를 보기위해서 groupby와 pivot_table을 사용
df.groupby('요일').mean() #요일로 그룹핑해서 각 수치형데이터의 각 컬렴별 평균을 반환
```

```python
df.groupby('요일').count() #각 컬럼별 빈도수(범주형 포함).데이터 프레임
#df.groupby('요일').size()#빈도수.시리즈
```

```python
df.groupby(['사고유형_대분류','사고유형_중분류'])['사망자수'].count()
```

```python
df.groupby(['사고유형_대분류','사고유형_중분류'])['사망자수'].agg(['count','mean','median'])
```

```python
#pivot_table()은 gruopby와 거의 같은 기능이나 직관적이다
#df.groupby('요일').count()
df.pivot_table(index='요일',aggfunc='count')
```

```python
#df.groupby(['사고유형_대분류','사고유형_중분류'])['사망자수'].count()
df.pivot_table(index=['사고유형_대분류','사고유형_중분류'],values='사망자수',aggfunc='count')
```

```python
#df.groupby(['사고유형_대분류','사고유형_중분류'])['사망자수'].agg(['count','mean','median'])
df.pivot_table(index=['사고유형_대분류','사고유형_중분류'],values='사망자수',aggfunc=['count','mean','median'])
```
